```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados
url = "https://raw.githubusercontent.com/higoramario/univesp-com400-visualizacao-computacional/main/semana-2-exercicio-apoio-2-campeoes-mundiais.csv"
data = pd.read_csv(url)

# Remover os anos de 1975 e 1978 e considerar os dois campeonatos de 2000
data = data[~data['ANO'].isin([1975, 1978])]
data['CAMPEÃO'] = data['CAMPEÃO'].replace('Corinthians', 'Corinthians (2000, 2000)')

# Contar os títulos por clube
title_count = data['CAMPEÃO'].value_counts()

# Criar o gráfico de barras
plt.figure(figsize=(12, 8))
title_count.plot(kind='bar', color='skyblue')
plt.title('Clubes Campeões do Campeonato Mundial Interclubes de Futebol Masculino')
plt.xlabel('Clubes Campeões')
plt.ylabel('Número de Títulos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
